```python
import numpy as np # for numerical operations
import pandas as pd # for handling input data
import matplotlib.pyplot as plt # for data visualization 
import seaborn as sns # for data visualization
train = pd.read_csv('BMI_Dataset_train.csv') # uses pandas library to open a .csv file
test = pd.read_csv('BMI_Dataset_test.csv')
```

```python
train.head()
```

```python
x1_train= train['Height(meter)']
x2_train= train['Weight(KG)']
label=train['BMI']
```

```python
x1=np.array(x1_train)
x2=np.array(x2_train)
y=np.array(label)
```

```python
y.shape
```

### Gradient Descent finds the parameters that minimize the cost function (error in prediction). 
depends on
1- Iterations
2- Learning rate


```python
w1=1
w2=1
b=1
alpha=0.0005 #between 0 and 1
for i in range (1000000):
#predict
    y1=w1*x1+w2*x2+b
    w=y1-y
#cost function#calculate error
    j=((np.sum(np.square(w)))/2000)
#learn function
    b=b-((alpha/1000)*np.sum(w))
    w1=w1-((alpha/1000)*np.sum(w*x1))
    w2=w2-((alpha/1000)*np.sum(w*x2))
    if i%100000:
        print("j :",j)
        
```
